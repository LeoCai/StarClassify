```python
import pandas as pd
from pandas import DataFrame
from pandas import concat
import matplotlib.pyplot as plt
import xgboost as xgb

%matplotlib inline
```

```python
test_index = pd.read_csv('./datas/first_test_index_20180131.csv')
```

```python
def parseFile(path, outPath,id):
    info = pd.read_csv(path+id+'.txt',header=None).loc[0];
    feature = [[ info.mean(), info.max(), info.min(), info.std()]]
    return pd.DataFrame(feature);
    #parsed = info.sum
    #parsed.to_csv('./datas/parsedFile/'+id+'.txt');
    
```

```python
inPath = './datas/first_test_data_20180131/'
outPath = './datas/test_data/'
features = pd.DataFrame()
finished = 0
total = test_index.size
for idx, row in test_index.iterrows():
    features.append(parseFile(inPath, outPath, str(row.id)))
    finished = finished + 1
    if finished%100 == 0:
        print('process:',finished, total, 1.0*finished/total)
```

```python
train_index = pd.read_csv('./datas/first_train_index_20180131.csv')
inPath = './datas/first_train_data_20180131/'
outPath = './datas/train_data/'
features_train = pd.DataFrame()
finished = 0
total = train_index.size
for idx, row in train_index.iterrows():
    features_train.append(parseFile(inPath, outPath, str(row.id)))
    finished = finished + 1
    if finished%100 == 0:
        print('process:',finished, total, 1.0*finished/total)
```
